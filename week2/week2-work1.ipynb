## 【説明課題】 統計基礎用語の説明

- **標準偏差について、分散との違いや関係性、どんな時に使う指標なのか**   
標準偏差および分散は、データのばらつき具合を示すときに用いる指標であり、小さいほどばらつきが少なく、大きいほどばらつきが大きいことを表す。分散は各データとその平均値の差の2乗を全て足し、その平均をとったものである。標準偏差は分散の平方根をとった値であり、データと同じ単位で表すことができる。また、モデルを組む前にデータの特徴を把握するために用いられる。   
$ n $ 個の確率変数がそれぞれ $ x^{}_{1}, x^{}_{2}, ... ,x^{}_{n}, $ という値を取り、平均値 $ \bar{x} $ のとき、分散 $ \sigma^2 $ は、
<br>
$ \sigma^2 = \frac{1}{n} \displaystyle{ \sum_{k=1}^{n} } (x^{}_{k} - \bar{x})^2 $
<br>
<br>
標準偏差 $ \sigma $ は、
<br>
$ \sigma = \sqrt{\sigma^2} = \sqrt{\frac{1}{n} \displaystyle{ \sum_{k=1}^{n} } (x^{}_{k} - \bar{x})^2} $
<br>
<br>
- **共分散行列について**   
共分散行列は、分散の概念を多次元確率変数に拡張して行列としたものであり、対角成分に分散が並び、非対角成分には共分散が並ぶ。   
確率変数が $ n $ 個の場合の共分散行列は、$ n $ 個の確率変数 $ X1, X2, ..., Xn $ に対して、 $ \sum $ を共分散行列とすると、第 $ ii $ 成分は$ V[X^{}_{i}] $ 、第 $ ij $ 成分 $ (i ≠ j) $ は $ Cov[X^{}_{i}, X^{}_{j}] $ であるような $ n × n $ 行列として定義される。($ V[X^{}_{i}] $は$ X^{}_{i} $ の分散、$ Cov[X^{}_{i}, X^{}_{j}] $は$ X^{}_{i}, X^{}_{j} $ の共分散)   
<br>
<br>
$ \sum = 
\begin{bmatrix}
V[X^{}_{1}] & Cov[X^{}_{1}, X^{}_{2}] & \ldots & Cov[X^{}_{1}, X^{}_{n}] \\\ 
Cov[X^{}_{2}, X^{}_{1}] & V[X^{}_{2}] & \ldots & Cov[X^{}_{2}] \\\
\vdots & \vdots & \ddots & \vdots \\\
Cov[X^{}_{n}, X^{}_{1}] & Cov[X^{}_{n}, X^{}_{2}] & \ldots & V[X^{}_{n}] \\\
\end{bmatrix}
$
<br>
<br>
- **1次元の正規分布と多次元の正規分布について述べよ。また、特にどういう時に使うのか**   
1次元正規分布とは、左右対称の連続型の確率分布であり、そのグラフは、平均値を中央にして、左右対称の釣鐘型となる。また、平均を $ \mu $ 、分散を $ \sigma^2 $ とした場合に、その確率密度関数は $ f(x) = \frac{1}{\sqrt{2}\pi \sigma} exp\left\{-\frac{(x - \mu)^2}{2\sigma^2}\right\} $ となる。平均±1σの範囲には全体の68.26% 、平均 ±2σの範囲には全体の95.44% 、平均±3σの範囲には全体の99.73% が収まる。   
多次元正規分布について、 平均 $ \mu $ の $ n $ 次元正規分布の同時確率密度関数は、 $ x = (x^{}_{1}, x^{}_{2}, …, x^{}_{n}) $ 、 $ \sum $ を共分散行列とすると
 $ f(\vec{x}) = \frac{1}{(\sqrt{2\pi})^n\sqrt{|\sum|}}  exp\left\{-\frac{1}{2}(\vec{x} - \vec{\mu})^T \sum^{-1}(\vec{x} - \vec{\mu})\right\} $ である。   
正規分布は、平均と分散が分かれば仮定することができる。そのため、未知の確率分布に対しても、正規分布と仮定して計算してみるということができる。また、分布の中心に近い値ほど観察されやすく、ある仮想的な範囲から値をランダムにサンプルする場合、サンプルの数を増やすほど正規分布に近づく。ランダムサンプリングを伴う調査において、調査で扱われる指標が正規分布をとると仮定すると、分布の中心から極端に遠い値は、外れ値として無視することができる。

## 【コーディング課題】 データ点の作成

```python
import numpy as np

#data1:中心が(-3, 0)のデータ500個
data1 = np.random.multivariate_normal([-3, 0], [[1.0, 0.8], [0.8, 1.0]], 500)

#出力
data1
```

```python
#data2:中心が(0, -3)のデータ500個
data2 = np.random.multivariate_normal([0, -3], [[1.0, 0.8], [0.8, 1.0]], 500)

#出力
data2
```

## 【コーディング課題】 データ点へのラベル付け

```python
#data1とdata2を結合
data = np.concatenate((data1, data2))

#dataの3列目に1を追加
data = np.insert(data[0:1000], 2, 1, axis=1)

#dataの501~1000行目、3列目を0に変換
data[500:1000, 2] = [0]

#出力
data
```

## 【コーディング課題】 正しくラベル付けされているか見てみる

```python
import matplotlib.pyplot as plt
%matplotlib inline

#dataの配列の要素数を取得
row, col = data.shape

#3列目が1の時黄色、0の時紫色の散布図
for i in range(row):
    if data[i, 2] == 1:
        color = 'yellow'
    else:
        color = 'purple'
    
    #出力
    plt.scatter(data[i, 0], data[i, 1], s=50, c=color, marker='o');
```
