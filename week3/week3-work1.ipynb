## 【課題6】 図の作成

データは可視化して確認することが大切です。図を作成していきましょう。Matplotlibやseabornを使ってください。

ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）

- 特徴量を一つ選び、boxplotを使ってラベルごとのデータの分布を可視化する
- 特徴量を一つ選び、violinplotを使ってラベルごとのデータの分布を可視化する

boxplotとviolinplotは似た図を表示します。これらはどう違い、それぞれどのような利点があるのか説明してください。



### ・ラベルごとのサンプル数を円グラフにする(パーセンテージまで表示する)

```python
%matplotlib inline
import matplotlib.pyplot as plt

#ラベルごとのサンプル数
species0 = len(df['Species'] == 0)
species1 = len(df['Species'] == 1)
species2 = len(df['Species'] == 2)

#ラベル名の読み込み
label = iris.target_names

#円グラフを表示
x = [species0, species1, species2]
plt.pie(x, labels=label, autopct='%.1f%%');
```

### ・boxplotを使いデータの分布を可視化

```python
import seaborn as sns

#縦軸に各特徴量、横軸に各ラベルのboxplot
for i in range(4):
    boxplot = sns.boxplot(x=df['Species'], y=df.iloc[:, i])
    boxplot.set_xticklabels(label)
    plt.show()
```

### ・violinplotを使いデータの分布を可視化

```python
#縦軸に各特徴量、横軸に各ラベルのviolinplot
for i in range(4):
    violinplot = sns.violinplot(x=df['Species'], y=df.iloc[:, i])
    violinplot.set_xticklabels(label)
    plt.show()
```

### ・boxplotとviolinplotの違いと利点
boxplotは真ん中の線が中央値、箱の上下は第一・第三四分位、ひげの上下は最大値・最小値である。外れ値は点で表される。
violinplotはデータの分布の密度を可視化しており、内部には箱ひげ図が描かれている。
boxplotはデータのばらつきを見ることができるが、violinplotはデータの分布密度も見ることができる。

## 【課題7】 特徴量間の関係の確認

特徴量間の関係を見ていきます。

- 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る
- 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る
- ４つの特徴量に関する相関係数行列を作る
- 相関係数行列のヒートマップを作る

### ・縦軸と横軸に特徴量を1つずつ取り、種類毎に色分けした散布図を作る

```python
#種類ごとに色分け
color = {0: 'red', 1: 'blue', 2: 'yellow'}
 
#縦軸、横軸に特徴量を1つずつとり出力
for j in range(4):
    for k in range(4):
        
        #特徴量が同じなら出力しない
        if j == k:
            pass
        
        #特徴量が違う場合は散布図を描写
        else:
            plt.xlabel(df.columns[k])
            plt.ylabel(df.columns[j])
            for i in range(len(df)):
                plt.scatter(df.iloc[i, k], df.iloc[i, j], c=color[df.iloc[i, 4]])
            plt.show()
```

### ・散布図行列を作る

```python
#dfをコピー
df_copy = df.copy()

#Speciesのラベルを番号からirisの名前に変更
df_copy[['Species']] = label[df_copy['Species']]

#散布図行列を出力
sns.pairplot(df_copy, vars=df.iloc[:, 0:4], hue='Species');
```

### ・4つの特徴量に関する相関係数行列を作る

```python
df_corr = df.iloc[:, 0:4].corr()
df_corr
```

### ・相関係数行列のヒートマップを作る

```python
sns.heatmap(df_corr);
```

setosaは花弁の長さ・広さがversicolor、virginicaに比べて小さい。また、versicolorとvirginicaの花弁の長さ・広さは重なっている部分もあるが、versicolorの方が小さい傾向にある。そのため、花弁の長さ、広さでの分類が可能と思われる。

萼片の長さはsetosa、versicolor、virginicaの順に小さい傾向が若干あるが、重なっている部分も多い。また、setosaの萼片の長さは他の2つ比べて比較的揃っている。

萼片の広さはどれも同じような大きさであり、分けることは難しいと思われる。
